```python

s = 0

counts = [1 for _ in range(205)]
points = [0 for _ in range(205)]
with open('input.txt', 'r') as f:
    t = f.read()
    for idx, line in enumerate(t.split('\n')):
        l = line.split(':')[1]
        winning = set(l.split('|')[0].strip().split(' '))
        round = l.split('|')[1].strip().split(' ')

        c = 0
        for x in round:
            if x!="" and x in winning:
                c+=1

        if c>0:
            points[idx] = 2**(c-1)
            for i in range(idx+1, idx+1+c):
                if i<205:
                    counts[i] += counts[idx]
        print(points)
        print(counts)
        print(f"{sorted(winning)} | {sorted(round)}: {c}, ")

print(sum(counts))
```
