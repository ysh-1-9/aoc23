```python
import sys
sys.setrecursionlimit(150000)

directions = {"|": [(1,0), (-1,0)],
              "-": [(0,1), (0,-1)],
              "L": [(-1,0), (0,1)],
              "J": [(-1,0), (0,-1)],
              "7": [(1,0), (0,-1)],
              "F": [(1,0), (0,1)]}

grid = []
with open("input.txt", "r") as f:
    l = f.read()
    for line in l.split('\n'):
        grid.append(line)

def neybs(y,x):
    n = []
    if y+1<len(grid):
        n.append([y+1,x])
    if x-1>0:
        n.append([y, x-1])
    if y-1>0:
        n.append([y-1, x])
    if x+1<len(grid[0]):
        n.append([y, x+1])

    return n

def valid(y,x):
    if y>0 and y<len(grid) and x>0 and x<len(grid[0]):
        return True
    return False

visited = [[None for _ in grid[0]] for _ in grid]
starting = None

backatstart = True
def visit(y,x, backatstart):
    dist = None
    for dy,dx in directions[grid[y][x]]:
        if visited[y+dy][x+dx] is not None:
            dist = visited[y+dy][x+dx]
            if dist == 0:
                if backatstart:
                    backatstart = False
                else:
                    for dy,dx in directions[grid[y][x]]:
                        if visited[y+dy][x+dx]!=0:
                            dist = visited[y+dy][x+dx]
                        visited[y][x] = dist+1
                        return
                    
    visited[y][x] = dist+1
    for dy,dx in directions[grid[y][x]]:
        if visited[y+dy][x+dx] is None:
            visit(y+dy,x+dx, backatstart)
            return

def start(y,x):
    for ny,nx in neybs(y,x):
        print(f"ny: {ny}, nx: {nx}, ch: {grid[ny][nx]}")
        for dy,dx in directions[grid[ny][nx]]:
            print(f"dy: {dy}, dx: {dx}")
            if ny+dy == y and nx+dx == x:
                print("here")
                visited[y][x] = 0
                visit(ny,nx, True)
                return

for y, line in enumerate(grid):
    for x, ch in enumerate(line):
        if ch=="S":
            starting = [y,x]
            print("starting at: ",y,x,ch)
            start(y,x)
            break
            
```

```python
def allneybs(y,x):
    ns = []
    diffs = [0,1,-1]
    for dy in diffs:
        for dx in diffs:
            if(valid(y+dy,x+dx) and not(dy==0 and dx==0)):
                ns.append([y+dy, x+dx])
    return ns

print(allneybs(107,110))
```

```python
grouped = [[None for _ in grid[0]] for _ in grid]

# does group i touch the wall?
walled = []
count = []
firstidx = []

def bfsvisit(y,x,group):
    print("visiting: ",y,x,group)
    walled = False
    
    ns = neybs(y,x)
    print("neybs: ", ns)
    if len(ns)<4:
        walled = True
    print("grouped: ", grouped[y][x])
    if grouped[y][x] is not None:
        return walled

    count[group]+=1
    grouped[y][x] = group
    for ny, nx in ns:
        if visited[ny][nx] is None:
            print("recursion")
            walled = walled or bfsvisit(ny,nx,group)
    return walled

next = 0
for y, line in enumerate(grid):
    for x, ch in enumerate(line):
        if visited[y][x] is None and grouped[y][x] is None:
            count.append(1)
            firstidx.append([y,x])
            walled.append(bfsvisit(y,x,next))
            next+=1
```

```python
def crossy(ch):
    if ch!="|" and ch!=".":
        return True

def crossx(ch):
    if ch!="-" and ch!=".":
        return True
        
def inside(y,x):
    things = []

    above = [grid[yi][x] for yi in range(y) if visited[yi][x] is not None]
    things += [sum(1 if crossy(ch) else 0 for ch in above)]

    
    if y+1!=len(grid):
        #print(list((yi,x) for yi in range(y+1,len(grid))))
        #below = 
        things+= [sum(1 if crossy(ch) else 0 for ch in (grid[yi][x] for yi in range(y+1,len(grid)) if visited[yi][x] is not None))]

    
    things += [sum(1 if crossx(ch) else 0 for ch in (grid[y][xi] for xi in range(x) if visited[y][xi] is not None))]
    
    if x!=len(grid[0])-1:
        things += [sum(1 if crossx(ch) else 0 for ch in (grid[y][xi] for xi in range(x+1,len(grid[0])) if visited[y][xi] is not None))]
    print(things)
    for t in things:
        if t%2==0:
            return False
    return True

s = 0
for y,line in enumerate(grid):
    for x,ch in enumerate(line):
        if visited[y][x] is None and inside(y,x):
            print(y,x)
            s+=1
print(s)
```
