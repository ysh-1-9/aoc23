```python
import re

digit = 'zero|one|two|three|four|five|six|seven|eight|nine|[0-9]'

def convert(s):
    try:
        r = int(s)
        return r
    except:
        list = enumerate(digit.split('|')[:-1])
        map = {x: idx for idx, x in list}
        return map[s]

s = 0
with open('input.txt', 'r') as f:
    t = f.read()
    for line in t.split('\n'):
        res = re.findall(digit, line)
        if len(res) < 2:
            print(f"{line}: {res[0]}")
            s+=convert(res[0])*11
        else:
            print(f"{line}: {res[0]}, {res[-1]}")
            s+=convert(res[0])*10 + convert(res[-1])
print(s)
```
