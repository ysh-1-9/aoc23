```python
from functools import cmp_to_key

card_order = ['J', '2', '3', '4', '5', '6', '7', '8', '9', 'T', 'Q', 'K', 'A']

def get_type(hand):
    #h = sorted(hand, key = lambda x: card_order.index(x))
    c = {}
    maxyet = 0
    maxc = None
    jc = 0
    for x in hand:
        if x!= 'J':
            c[x] = c.get(x, 0)+1
            if c[x] > maxyet:
                maxyet = c[x]
                maxc = x
        else:
            jc+=1

    
    maxyet+=jc
    if maxc:
        c[maxc] = maxyet
    else:
        return 1

    if maxyet == 5:
        return 1
    elif maxyet == 4:
        return 2
    elif maxyet == 3:
        if 2 in c.values():
            return 3
        return 4
    elif maxyet == 2:
        c2 = 0
        for x in c.values():
            if x==2:
                c2+=1
        if c2==2:
            return 5
        return 6
    else:
        return 7
    

def compare(firsty, secondy):
    first = firsty[0]
    second = secondy[0]
    
    tf = get_type(first)
    ts = get_type(second)
    
    if tf<ts:
        return 1
    elif ts<tf:
        return -1

    for x,y in zip(first, second):
        tx = card_order.index(x)
        ty = card_order.index(y)
        if tx>ty:
            return 1
        elif ty>tx:
            return -1

    return 0

s = 0
with open("input.txt", "r") as f:
    l = f.read()
    input = [line.strip().split() for line in l.split('\n')]
    output = sorted(input,key=cmp_to_key(compare))
    #print(output)
    print(len(output))
    for idx,x in enumerate(output):
        print(x)
        print(f"{idx+1}*{x[1]}")
        s+=(idx+1)*int(x[1])
    print(s)
    print(sum((idx+1)*int(bid) for idx,(_,bid) in enumerate(output)))
```
